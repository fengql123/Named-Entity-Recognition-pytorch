```python
import pandas as pd
import numpy as np

data = pd.read_csv("./data/ner_datasetreference.csv", encoding='latin1')
data
```

```python
def getData(data):
    data = data.fillna(method = 'ffill')
    data = data.drop(['POS'], axis = 1)
    sentences = []
    labels = []
    agg_function = lambda s: [(w,t) for w,t in zip(s["Word"].values.tolist(),
                                                        s["Tag"].values.tolist())]
    group = data.groupby('Sentence #').apply(agg_function)
    sentence = [s for s in group]
    return [' '.join([word[0] for word in s]) for s in sentence], [[lab[0] for lab in s] for s in sentence]

x,y = getData(data)
x[:100]
```
